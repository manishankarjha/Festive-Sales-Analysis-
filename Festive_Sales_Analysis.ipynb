# Exploratory Data Analysis

### Gender

```python
# plotting a bar chart for Gender and it's count

ax = sns.countplot(x = 'Gender',data = df)      #this plot object is stored in variable ax

ax.containers
# The plot object has a method called containers that would list the properties of each bar. 
```

```python
ax = sns.countplot(x = 'Gender',data = df) 

# Iterate through the list items of the container object and pass each item to the bar_label function. 
# This will extract and display the bar value in the bar plot.

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
# plotting a bar chart for total amount vs gender 

sales_gen = df.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
sales_gen
```

```python
sns.barplot(x = 'Gender',y= 'Amount' ,data = sales_gen)
```

*From above graphs we can see that most of the buyers are females and even the purchasing power of females are greater than men*

### Age

```python
ax = sns.countplot(data = df.sort_values('Age Group'), x = 'Age Group', hue = 'Gender')

# df.sort_values('Age Group') is a new dataframe where the age groups have been sorted in ascending(by default) order

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
# Total Amount vs Age Group
sales_age = df.groupby(['Age Group'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.barplot(x = 'Age Group',y= 'Amount' ,data = sales_age)
```

*From above graphs we can see that most of the buyers are of age group between 26-35 yrs female*

### State

```python
# total number of orders from top 10 states

sales_state = df.groupby(['State'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)

#setting figure size by passing dictionary to rc parameter with key 'figure.figsize' in seaborn set method:

sns.set(rc={'figure.figsize':(15,5)})               #figure size in inches

sns.barplot(data = sales_state, x = 'State',y= 'Orders')
```

**Other alternative may be to use ``figure.figsize`` of ``rcParams`` to set figure size as below:**

``from matplotlib import rcParams``

``rcParams['figure.figsize'] = 12,8``                           

The no. of orders from Kerala are more in number but not in amount than Haryana as can be seen below:

```python
# total amount/sales from top 10 states

sales_state = df.groupby(['State'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)

sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(data = sales_state, x = 'State',y= 'Amount')
```

*From above graphs we can see that most of the orders & total sales/amount are from Uttar Pradesh, Maharashtra and Karnataka respectively*


### Marital Status

```python
ax = sns.countplot(data = df, x = 'Marital_Status')

sns.set(rc={'figure.figsize':(7,5)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
# Total Amount vs Marital Status
sales_ms = df.groupby(['Marital_Status', 'Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.set(rc={'figure.figsize':(6,5)})
sns.barplot(data = sales_ms, x = 'Marital_Status',y= 'Amount', hue='Gender')
# As in hue we want Gender column also, that's why we have taken it also while making a new dataframe during grouping above
```

*From above graphs we can see that most of the buyers are married (women) and they have high purchasing power*

### Occupation

```python
a=df['Occupation'].value_counts()
pd.DataFrame(data=a)
```

```python
a=df['Occupation'].value_counts()
df1 = pd.DataFrame(data=a)
df1['Name'] = df1.index
df1.set_index(np.arange(len(a)),inplace=True)
df1
```

```python
df2 = df.groupby('Occupation', as_index=False).count()[['Occupation','User_ID']]
df2.sort_values(by='User_ID',ascending=False)
```

```python
large_to_small = df.groupby('Occupation').size().sort_values()   #size() will count occurences by groups
large_to_small
```

```python
large_to_small.index[::-1]               #It will take all indices from 0 to last [ : ] and then reverses the order[ : :-1]
```

```python
sns.set(rc={'figure.figsize':(20,5)})

ax = sns.countplot(data = df, x = 'Occupation',order=large_to_small)  #we want only indices name in the order argument

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
# Total Amount vs Occupation
sales_occ = df.groupby(['Occupation'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data = sales_occ, x = 'Occupation',y= 'Amount')
```

*From above graphs we can see that most of the buyers are working in IT, Healthcare and Aviation sector*

### Product Category

```python
sns.set(rc={'figure.figsize':(24,5)})
ax = sns.countplot(data = df, x = 'Product_Category')

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
# Total Amount vs Product_Category 
sales_pc = df.groupby(['Product_Category'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)

sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data = sales_pc, x = 'Product_Category',y= 'Amount')
```

*From above graphs we can see that most of the sold products are from Food, Clothing and Electronics category*

```python
# Orders vs Product_ID
sales_pi = df.groupby(['Product_ID'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)

sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data = sales_pi, x = 'Product_ID',y= 'Orders')
```

```python
# top 10 most sold products (same thing as above)

fig1, ax1 = plt.subplots(figsize=(12,7))
df.groupby('Product_ID')['Orders'].sum().nlargest(10).sort_values(ascending=False).plot(kind='bar')

# Note- Here we do not have to give as_index=False so as to create a new index column as we are plotting with plot function 
# which automatically plots the required variables according to their nature.
```
